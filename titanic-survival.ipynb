# Prediction de la survie dans Titanic

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
# load data from seaborn
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
data = titanic.copy()
```

```python
data = data[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch','alone']]
data
```

```python
# drop NaN values
data.dropna(axis=0, inplace = True)

# convert 'sex' values to 0(male) and 1(female)
data['sex'].replace(['male', 'female'], [0, 1], inplace=True)

# convert 'alone' values to 0(False) and 1(True)
data['alone'] = data['alone'].astype(int)
data
```

```python
X = data.drop('survived', axis = 1)
y = data['survived']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Find the best model parameters using GridSearchCV
param_grid = {'n_neighbors':np.arange(1, 100), 'metric':['euclidean', 'manhattan']}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv = 5)
grid.fit(X_train, y_train)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
model = grid.best_estimator_
model.score(X_test, y_test)
```

```python
def survival(model, pclass=3, sex=0, age=18, sibsp=0, parch=0, alone=0):
    X = np.array([pclass, sex, age, sibsp, parch, alone]).reshape(1, 6)
    print(model.predict(X))
    print(model.predict_proba(X))
```

```python
# Prediction here
# Result : 1 survived, 0: died
survival(model, pclass=3, sex=0, age=10, alone=1)
```
